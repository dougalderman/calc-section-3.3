```r
x<-c(10,11,12,13,14,15)
y<-c(280,294,309,324,350,406)
```

```r
plot(x,y,col="blue", main="Estimated Paper Shredder Sales",xlab="# of Years Since 1990", ylab="Paper Shredder Sales (Millions of Dollars)")
```

```r
relation_nl<-nls(y~I(c*exp(k*x)), start=list(c=130,k=0.06))
print(relation_nl)
summary(relation_nl)
```

```r
x_seq <- seq(from=10,to=15, by=0.1)
new_relation_nl<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation_nl[i]<-predict(relation_nl, x_new)
}
plot(x,y,col="blue", main="Nonlinear Regression Example",lines(x_seq, new_relation_nl),xlab="x",ylab="y")
```

```r
pred_17 = predict(relation_nl, x_17)
```
